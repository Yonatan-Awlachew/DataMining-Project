# Data Preparation

Importing Pandas Library

```python
import pandas as pd
```


## Economic Data droping other Colums and Spliting into 6 datasets from 1980-2023


### Check if there is Duplicate Value

```python
df = pd.read_csv('Economic accounts for agriculture - values atconstant prices_combined_1980_2023.csv')

duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
  print("Duplicate rows found:")
  print(duplicate_rows)
else:
  print("No duplicate rows found in the merged DataFrame.")
```

### Overall Agricultural Economic Data

```python
df = pd.read_csv('Economic accounts for agriculture - values atconstant prices_combined_1980_2023.csv')
df.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq','indic_ag', 'unit','geo','OBS_FLAG'], inplace=True)


Economic_Data = df[df['itm_newa'] == 16000]
Economic_Data = Economic_Data.rename(columns={'OBS_VALUE': 'Agriculture_Economy(Million euro)'})
Economic_Data=Economic_Data.drop('itm_newa',axis=1)
Economic_Data=Economic_Data.reset_index(drop=True)

Economic_Data.head()
```

###Crop_OutPut prices

```python
Crop_OutPut = df[df['itm_newa'] == 10000]
Crop_OutPut = Crop_OutPut.rename(columns={'OBS_VALUE': 'Crop_Economy(Million euro)'})
Crop_OutPut=Crop_OutPut.drop('itm_newa',axis=1)
Crop_OutPut=Crop_OutPut.reset_index(drop=True)
Crop_OutPut.head()
```

### Milk Products Overall Income


```python
Milk_Overall = df[df['itm_newa'] == 12100]
Milk_Overall = Milk_Overall.rename(columns={'OBS_VALUE': 'Milk_Economy(Million euro)'})
Milk_Overall=Milk_Overall.drop('itm_newa',axis=1)
Milk_Overall=Milk_Overall.reset_index(drop=True)
Milk_Overall.head()
```

### Animal Products Overall Income


```python
Animal_Overall = df[df['itm_newa'] == 13000]
Animal_Overall = Animal_Overall.rename(columns={'OBS_VALUE': 'Animal_Economy(Million euro)'})
Animal_Overall=Animal_Overall.drop('itm_newa',axis=1)
Animal_Overall=Animal_Overall.reset_index(drop=True)
Animal_Overall.head()
```

### Fertilisers and soil improver Cost

```python
Fertilisers_Cost = df[df['itm_newa'] == 19030]
Fertilisers_Cost = Fertilisers_Cost.rename(columns={'OBS_VALUE': 'Fertilisers_Cost(Million euro)'})
Fertilisers_Cost=Fertilisers_Cost.drop('itm_newa',axis=1)
Fertilisers_Cost=Fertilisers_Cost.reset_index(drop=True)
Fertilisers_Cost.head()
```

### Plant protection products, herbicides, insecticides and pesticides Costs

```python
Plant_protection_Cost = df[df['itm_newa'] == 19040]
Plant_protection_Cost = Plant_protection_Cost.rename(columns={'OBS_VALUE': 'Plant_Protection_Cost(Million euro)'})
Plant_protection_Cost=Plant_protection_Cost.drop('itm_newa',axis=1)
Plant_protection_Cost=Plant_protection_Cost.reset_index(drop=True)
Plant_protection_Cost.head()
```

Economic accounts for agriculture Subsidies on products1980_2023

```python
Agriculture_Subsidies = pd.read_csv('Economic accounts for agriculture Subsidies on products1980_2023.csv')
Agriculture_Subsidies.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'itm_newa','indic_ag','unit','geo','OBS_FLAG'], inplace=True)

Agriculture_Subsidies = Agriculture_Subsidies.rename(columns={'OBS_VALUE': 'Agriculture_Subsidies(Million euro)'})

Agriculture_Subsidies.head()
```

### All the 7 dataframe combined

```python
dataframes = [Crop_OutPut, Milk_Overall, Animal_Overall, Fertilisers_Cost, Plant_protection_Cost,Agriculture_Subsidies,Economic_Data]

merged_df = dataframes[0]

for df in dataframes[1:]:
    merged_df = pd.merge(merged_df, df, on='TIME_PERIOD', how='inner')
merged_df.head()
```

## Crop production droping other Colums and Concatinating the 2 datasets from 1980-2023

```python
df1 = pd.read_csv('Crop production - historical data(1980-1999).csv')
df1.head()
df1.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'strucpro','geo','OBS_FLAG'], inplace=True)

df2 = pd.read_csv('Crop production in EU standard humidity 2000-2023.csv')
df2.head()
df2.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'strucpro','geo','OBS_FLAG'], inplace=True)


x=pd.concat([df1,df2])

x.head()
```

### Cereals for the production of grain (including seed)

```python
Cereals_Production = x[x['crops'] == 'C0000']
Cereals_Production = Cereals_Production.rename(columns={'OBS_VALUE': 'Cereals_Production(1000 ha)'})
Cereals_Production=Cereals_Production.drop('crops',axis=1)
Cereals_Production=Cereals_Production.reset_index(drop=True)
Cereals_Production.head()
```

### Dry pulses and protein crops for the production of grain (including seed and mixtures of cereals and pulses)


```python
DryPulses_Production = x[x['crops'] == 'P0000']
DryPulses_Production = DryPulses_Production.rename(columns={'OBS_VALUE': 'DryPulses_Production(1000 ha)'})
DryPulses_Production=DryPulses_Production.drop('crops',axis=1)
DryPulses_Production=DryPulses_Production.reset_index(drop=True)
DryPulses_Production.head()
```

###Combined of 2 Columns and new dataframe with overall Columns



```python
merged_df1=DryPulses_Production.merge(Cereals_Production, how='inner', on='TIME_PERIOD')
merged_df1['Crop_Production(1000 ha)'] = merged_df1['Cereals_Production(1000 ha)'] + merged_df1['DryPulses_Production(1000 ha)']
merged_df1.head()
Crop_Production=merged_df1.drop(['Cereals_Production(1000 ha)','DryPulses_Production(1000 ha)'],axis=1)
Crop_Production.head()
```

## Meat production droping other Colums and final dataset from 1980-2023

```python
df3 = pd.read_csv('Meat production and foreign trade - head -monthly data 1980-2023.csv')
df3.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'meatitem', 'unit','geo','OBS_FLAG'], inplace=True)
df3
```

### Bovine meat Production in Thousand tonnes


```python
Bovine_Meat = df3[df3['meat'] == 'B1000']
Bovine_Meat = Bovine_Meat.rename(columns={'OBS_VALUE': 'Bovine_Meat_Production(1000 tonnes)'})
Bovine_Meat=Bovine_Meat.drop('meat',axis=1)
Bovine_Meat=Bovine_Meat.reset_index(drop=True)
Bovine_Meat.head()
```

### Meat of sheep and goats Production

```python
SheepGoat_Meat = df3[df3['meat'] == 'B4000']
SheepGoat_Meat = SheepGoat_Meat.rename(columns={'OBS_VALUE': 'SheepGoat_Meat_Production(1000 tonnes)'})
SheepGoat_Meat=SheepGoat_Meat.drop('meat',axis=1)
SheepGoat_Meat=SheepGoat_Meat.reset_index(drop=True)
SheepGoat_Meat.head()
```

###Combined of 2 Columns and new dataframe with overall Columns


```python
merged_df2=SheepGoat_Meat.merge(Bovine_Meat, how='inner', on='TIME_PERIOD')
merged_df2['Meat_Production(1000 tonnes)'] = merged_df2['SheepGoat_Meat_Production(1000 tonnes)'] + merged_df2['Bovine_Meat_Production(1000 tonnes)']
merged_df2.head()
Meat_Production=merged_df2.drop(['SheepGoat_Meat_Production(1000 tonnes)','Bovine_Meat_Production(1000 tonnes)'],axis=1)
Meat_Production.head()
```

## Raw milk, total available on dairies droping other Colums from 1980-2023

```python
Milk_Production = pd.read_csv('Milk collection (all milks) and dairy productsobtained - annual data 1980-2023.csv')
Milk_Production.drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'dairyprod','milkitem','geo','OBS_FLAG'], inplace=True)

Milk_Production = Milk_Production.rename(columns={'OBS_VALUE': 'Milk_Production(1000 tonnes)'})

Milk_Production.head()
```

## **Final Combining all Dataframes and Output of Combined DataFrame**

```python
dataframes1 = [Crop_Production,Meat_Production,Milk_Production, Crop_OutPut, Milk_Overall, Animal_Overall, Fertilisers_Cost, Plant_protection_Cost,Agriculture_Subsidies,Economic_Data]
merged_all1 = dataframes1[0]

for y in dataframes1[1:]:
    merged_all1 = pd.merge(merged_all1, y, on='TIME_PERIOD', how='inner')

merged_all1.shape

merged_all1.to_csv('Cleaned_dataset.csv', index=False)
```
